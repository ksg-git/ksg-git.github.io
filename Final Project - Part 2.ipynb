#### Code

```python
#--Grid heat map----------------------------------------------------------------------------
col_sc = bqplot.ColorScale(scheme="RdPu")
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()

c_ax = bqplot.ColorAxis(scale = col_sc, 
                        orientation = 'vertical', 
                        side = 'right')
x_ax = bqplot.Axis(scale = x_sc)
y_ax = bqplot.Axis(scale = y_sc, 
                   orientation = 'vertical')


heat_map = bqplot.GridHeatMap(color = np.log10(covidus),
                              row = covidus.index,
                              scales = {'color': col_sc,
                                        'row': y_sc,
                                        'column': x_sc},
                              interactions = {'click': 'select'},
                              anchor_style = {'fill':'blue'}, 
                              selected_style = {'opacity': 1.0},
                              unselected_style = {'opacity': 1.0})

#fig = bqplot.Figure(marks = [heat_map], axes = [c_ax, y_ax, x_ax])
#fig
```

```python
#--Line graph----------------------------------------------------------------------------
x_sch = bqplot.OrdinalScale()
y_sch = bqplot.LinearScale()
x_axh = bqplot.Axis(scale=x_sch, label='County')
y_axh = bqplot.Axis(scale=y_sch, label='Deaths', orientation='vertical')

#--Mask----------------------------------------------------------------------------
i,j = 16,5

mask = (temp2['state'] == covidus.index[i]) & \
       (temp2['NCHS_urbanization'] == covidus.columns.levels[1][j])

covidmask = temp2[mask]

grouped = covidmask.groupby("county_name")["deaths"].sum()
county = grouped.index
deaths = grouped.values

lines = bqplot.Lines(x = county, y = deaths,
                    scales = {'x': x_sch, 'y': y_sch})

#hist = bqplot.Hist(sample=grouped[i,j], normalized=False, # counts in each bin
#                  scales={'sample':x_sch, 'count':y_sch}, bins=5)


#grouped
#fig_lines = bqplot.Figure(marks = [lines], axes = [x_axh, y_axh])
#fig_lines
```

```python
#--On select function----------------------------------------------------------------------------
def on_selected(change):
    i,j = change['owner'].selected[0]
    mask = (temp2['state'] == covidus.index[i]) & \
           (temp2['NCHS_urbanization'] == covidus.columns.levels[1][j])

    covidmask = temp2[mask]

    grouped = covidmask.groupby("county_name")["deaths"].sum()
    county = grouped.index
    deaths = grouped.values
    lines.x = county
    lines.y = deaths

    
heat_map.observe(on_selected, 'selected')


#--Dashboard----------------------------------------------------------------------------

fig =  bqplot.Figure(marks = [heat_map], axes = [c_ax, y_ax, x_ax])
fig_lines = bqplot.Figure(marks = [lines], axes = [x_axh, y_axh])

fig.layout.min_width='500px'
fig.layout.min_height='900px'
fig_lines.layout.min_width='500px'

myDashboard = ipywidgets.HBox([fig, fig_lines])
myDashboard
```

#### Description for viewer

Left side of the dashboard lists the states and cities grouped according to the classification of the size and population density of the city, such as metropolitan, micropolitan, and the like. The color scale presents the confirmed covid-19 cases per 100000 people and the color scale is provided as well. Upon selecting one of the grids, right hand side graph will present the counties containing those type of cities and the deaths occurred from that county.

The relationship demonstrated by the visualization shows how as the cities grow larger there are more concentration of confirmed cases and, which counties containing certain cities were victim to the most damage due to lack of resources, the average age of the population, and the like. There certainly area lot more aspects to this visualization to be polished, however, such as labels. Just be aware that uncolored area(black area) do not contain data and may cause error upon selection.
